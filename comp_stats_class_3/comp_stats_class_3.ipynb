# Hipótese em Testes de Hipótese

## O que é uma Hipótese?

Uma **hipótese** é uma afirmação sobre um parâmetro populacional que desejamos verificar com base em evidências amostrais. Trabalhamos sempre com **duas hipóteses complementares**.

---

## As Duas Hipóteses

### H₀ — Hipótese Nula

- Representa o **status quo** — a ausência de efeito ou diferença.
- Sempre formulada com sinal de **igualdade** (`=`, `≤`, `≥`).
- É a hipótese que **tentamos rejeitar**.

**Exemplo:**
```
H₀: μ = 170
"A média de altura dos estudantes é igual a 170 cm."
```

---

### H₁ — Hipótese Alternativa

- Representa a **mudança ou efeito** que o pesquisador suspeita existir.
- Nunca contém igualdade — usa `≠`, `<` ou `>`.
- É o que buscamos **evidenciar com os dados**.

**Exemplo:**
```
H₁: μ ≠ 170
"A média de altura dos estudantes é diferente de 170 cm."
```

---

## Tipos de Teste

| Tipo | Símbolo em H₁ | Denominação |
|---|---|---|
| Bilateral | `≠` | Teste de duas caudas |
| Unilateral direito | `>` | Teste de cauda direita |
| Unilateral esquerdo | `<` | Teste de cauda esquerda |

---

## Erros de Decisão

Como decidimos com base em amostras, dois erros são possíveis:

| Decisão \ Realidade | H₀ **verdadeira** | H₀ **falsa** |
|---|---|---|
| **Rejeitar H₀** | ❌ Erro Tipo I `(α)` | ✅ Correto |
| **Não rejeitar H₀** | ✅ Correto | ❌ Erro Tipo II `(β)` |

- **Erro Tipo I (α):** Rejeitar H₀ quando ela é verdadeira → *falso positivo*
- **Erro Tipo II (β):** Não rejeitar H₀ quando ela é falsa → *falso negativo*

---

## Nível de Significância (α)

É a **probabilidade máxima tolerada de cometer o Erro Tipo I**.

| Valor de α | Rigor | Uso comum |
|---|---|---|
| `0,01` | Alto | Medicina, farmácia |
| `0,05` | Moderado | Ciências em geral |
| `0,10` | Baixo | Estudos exploratórios |

---

## O p-valor

O **p-valor** é a probabilidade de se obter um resultado tão extremo quanto o observado, *assumindo que H₀ é verdadeira*.

```
p-valor < α  →  Rejeita-se H₀  →  Resultado estatisticamente significativo
p-valor ≥ α  →  Não se rejeita H₀  →  Evidência insuficiente
```

> ⚠️ **Atenção:** "Não rejeitar H₀" **não prova** que ela é verdadeira.
> Significa apenas que os dados não forneceram evidência suficiente para refutá-la.

---

## Fluxo do Teste de Hipótese

```
┌─────────────────────────────────────┐
│  1. Formular H₀ e H₁               │
│  2. Definir o nível α               │
│  3. Coletar dados                   │
│  4. Calcular a estatística de teste │
│  5. Obter o p-valor                 │
│  6. Comparar p-valor com α          │
│  7. Concluir no contexto            │
└─────────────────────────────────────┘
```

---

## Exemplo Completo

> Uma empresa afirma que seus produtos duram em média **500 horas**.
> Um pesquisador suspeita que a durabilidade é **menor**.

| Etapa | Resultado |
|---|---|
| H₀ | μ = 500 horas |
| H₁ | μ < 500 horas (teste unilateral esquerdo) |
| α | 0,05 |
| p-valor calculado | 0,03 |
| Decisão | `0,03 < 0,05` → **Rejeita-se H₀** |
| Conclusão | Há evidência de que a durabilidade média é inferior a 500 horas. |

## Exercício 1: Amostragem Repetida — Efeito de N e S nas Distribuições Amostrais

---

## 1. Contexto do Problema

Temos uma **população** de $P$ elementos (inteiros de 1 a 100). Queremos entender o que acontece quando:

- Repetimos o sorteio de $S$ elementos **N vezes**
- Calculamos a **média** $\bar{x}$ de cada sorteio
- Observamos como essa coleção de médias se comporta

---

## 2. Os Dois Parâmetros que Controlam Tudo

| Parâmetro | Símbolo | O que controla |
|---|---|---|
| Número de repetições | $N$ | A **qualidade visual** do histograma |
| Tamanho da amostra | $S$ | A **precisão estatística** das médias |

> ⚠️ São efeitos **independentes** — um não cancela o outro.

---

## 3. Efeito de N — Qualidade do Histograma

### O que acontece?

Cada vez que sorteamos $S$ elementos e calculamos $\bar{x}$, obtemos **um ponto** no histograma. Quanto mais pontos (maior $N$), mais nítida e suave fica a forma da distribuição.

### Analogia

Imagine fotografar uma escultura com resolução crescente:

| $N$ | Resultado |
|---|---|
| 5 | Imagem borrada — mal dá para ver a forma |
| 50 | Forma reconhecível, mas com muito ruído |
| 200 | Contornos visíveis |
| 1.000 | Imagem nítida |
| 5.000 | Alta definição — curva normal evidente |

### Por que converge para a Normal?

Pelo **Teorema Central do Limite (TCL):**

> Independentemente da distribuição da população, a distribuição das médias amostrais converge para uma **distribuição normal** conforme $N$ cresce.

$$\bar{X} \sim \mathcal{N}\left(\mu,\ \frac{\sigma^2}{S}\right) \quad \text{quando } N \to \infty$$

---

## 4. Efeito de S — Precisão das Médias

### O que acontece?

O tamanho da amostra $S$ controla o **Erro Padrão (EP)** — a dispersão das médias em torno de $\mu$:

$$EP = \frac{\sigma}{\sqrt{S}}$$

### Consequência direta

| $S$ | $EP = \sigma / \sqrt{S}$ | Histograma |
|---|---|---|
| 2 | $28.87 / \sqrt{2} \approx 20.4$ | Muito largo e impreciso |
| 5 | $28.87 / \sqrt{5} \approx 12.9$ | Largo |
| 10 | $28.87 / \sqrt{10} \approx 9.1$ | Moderado |
| 20 | $28.87 / \sqrt{20} \approx 6.5$ | Estreito |
| 30 | $28.87 / \sqrt{30} \approx 5.3$ | Mais estreito |
| 50 | $28.87 / \sqrt{50} \approx 4.1$ | Muito estreito e preciso |

> Quanto maior $S$, mais as médias amostrais se **concentram em torno de $\mu$**.

### Por que $\sqrt{S}$ e não $S$?

Porque ao aumentar a amostra, os erros individuais se **cancelam parcialmente** entre si — o ganho de precisão segue uma lei de raiz quadrada, não linear. Dobrar $S$ reduz o EP por um fator de $\sqrt{2} \approx 1.41$, não por 2.

---

## 5. Os Dois Efeitos Combinados

```
         N pequeno                    N grande
         ┌──────────────┐             ┌──────────────┐
S pequeno│ Histograma   │             │ Histograma   │
         │ ruidoso e    │             │ suave e      │
         │ largo        │             │ largo        │
         └──────────────┘             └──────────────┘

         ┌──────────────┐             ┌──────────────┐
S grande │ Histograma   │             │ Histograma   │
         │ ruidoso e    │             │ suave e      │
         │ estreito     │             │ estreito ✓   │
         └──────────────┘             └──────────────┘
```

O ideal é **N grande + S grande** — histograma suave e estreito, centrado em $\mu$.

---

## 6. Lei dos Grandes Números

Além da forma do histograma, $N$ também controla a **convergência da média acumulada**:

$$\bar{X}_n = \frac{1}{n}\sum_{i=1}^{n} \bar{x}_i \xrightarrow{n \to \infty} \mu$$

Com poucos sorteios a média acumulada oscila bastante. Com muitos sorteios ela se estabiliza em torno de $\mu = 50.5$ — isso é a **Lei dos Grandes Números** em ação.

---

## 7. O Boxplot Conta a História de S

No boxplot, cada caixa representa a distribuição das $N$ médias para um valor de $S$:

| Elemento do Boxplot | O que representa |
|---|---|
| Linha vermelha central | Mediana das médias amostrais |
| Altura da caixa (IQR) | Dispersão central — proporcional ao $EP$ |
| Bigodes | Amplitude geral das médias |
| Pontos fora | Médias extremas (outliers) |

Conforme $S$ cresce, a **caixa encolhe** — evidência visual direta de que o $EP = \sigma/\sqrt{S}$ está diminuindo.

---

## 8. Resumo Final

```
Pergunta                          Resposta
─────────────────────────────────────────────────────────
Por que aumentar N?       → Histograma mais suave e normal
Por que aumentar S?       → Médias mais precisas e concentradas
O que é o Erro Padrão?    → EP = σ/√S — dispersão das médias
O que é o TCL?            → Médias → Normal quando N → ∞
O que é a Lei dos GN?     → Média acumulada → μ quando N → ∞
Por que √S e não S?       → Erros se cancelam parcialmente
```

> **Conclusão:** $N$ determina *o quanto você confia no que vê*. $S$ determina *o quanto suas estimativas são precisas*. Para uma boa análise estatística, você precisa dos dois em níveis adequados.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)

# ══════════════════════════════════════════════════════════
# População: inteiros de 1 a 100
# ══════════════════════════════════════════════════════════
P         = 100
populacao = np.arange(1, P + 1)
media_pop = populacao.mean()   # 50.5
std_pop   = populacao.std()    # ~28.87

# ══════════════════════════════════════════════════════════
# Configurações dos experimentos
# ══════════════════════════════════════════════════════════
Ns_plot = [5, 20, 100, 500, 2000, 5000]   # Variando N (S fixo)
Ss_plot = [2, 5, 10, 20, 30, 50]          # Variando S (N fixo)
S_fixo  = 10
N_fixo  = 1000

def sortear(P, S, N, populacao):
    """Realiza N sorteios de S elementos e retorna médias e desvios."""
    medias  = np.array([np.random.choice(populacao, size=S,
                        replace=False).mean() for _ in range(N)])
    desvios = np.array([np.random.choice(populacao, size=S,
                        replace=False).std(ddof=1) for _ in range(N)])
    return medias, desvios

# ══════════════════════════════════════════════════════════
# FIGURA 1 — Efeito de N (S fixo = 10)
# ══════════════════════════════════════════════════════════
fig1, axes1 = plt.subplots(2, 3, figsize=(18, 10))
fig1.suptitle(f'EFEITO DE N — Quanto mais repetições, mais a forma normal se revela\n'
              f'(S fixo = {S_fixo}, População = inteiros de 1 a {P})',
              fontsize=13, fontweight='bold')

ep = std_pop / np.sqrt(S_fixo)
x_range = np.linspace(media_pop - 4*ep, media_pop + 4*ep, 300)
curva_normal = norm.pdf(x_range, media_pop, ep)

cores_n = ['#EF9A9A', '#F48FB1', '#CE93D8', '#90CAF9', '#80CBC4', '#A5D6A7']

for idx, (N_val, cor, ax) in enumerate(zip(Ns_plot, cores_n, axes1.flat)):
    medias, _ = sortear(P, S_fixo, N_val, populacao)

    # Histograma
    ax.hist(medias, bins=max(10, N_val // 20), density=True,
            color=cor, edgecolor='white', linewidth=0.5,
            alpha=0.85, label=f'N={N_val} amostras')

    # Curva normal teórica
    ax.plot(x_range, curva_normal, color='#B71C1C',
            linewidth=2.2, linestyle='--', label='Normal teórica')

    # Linha da média
    ax.axvline(media_pop, color='#1A237E', linewidth=1.8,
               linestyle=':', label=f'μ = {media_pop:.1f}')

    # Média observada
    ax.axvline(medias.mean(), color='black', linewidth=1.2,
               linestyle='-', alpha=0.6,
               label=f'x̄ obs = {medias.mean():.2f}')

    ax.set_title(f'N = {N_val:,} repetições', fontsize=12, fontweight='bold')
    ax.set_xlabel('Média amostral $\\bar{x}$', fontsize=10)
    ax.set_ylabel('Densidade', fontsize=10)
    ax.legend(fontsize=8)
    ax.set_xlim(media_pop - 4*ep, media_pop + 4*ep)
    ax.spines[['top', 'right']].set_visible(False)

    # Anotação de desvio da normal
    residuo = abs(medias.mean() - media_pop)
    ax.text(0.97, 0.95, f'|x̄ − μ| = {residuo:.2f}',
            transform=ax.transAxes, fontsize=8,
            ha='right', va='top',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white',
                      edgecolor='gray', alpha=0.8))

plt.tight_layout()
plt.show()

# ══════════════════════════════════════════════════════════
# FIGURA 2 — Efeito de S (N fixo = 1000)
# ══════════════════════════════════════════════════════════
fig2, axes2 = plt.subplots(2, 3, figsize=(18, 10))
fig2.suptitle(f'EFEITO DE S — Quanto maior a amostra, mais estreita e precisa a distribuição\n'
              f'(N fixo = {N_fixo}, População = inteiros de 1 a {P})',
              fontsize=13, fontweight='bold')

cores_s = ['#EF9A9A', '#F48FB1', '#CE93D8', '#90CAF9', '#80CBC4', '#A5D6A7']

for idx, (S_val, cor, ax) in enumerate(zip(Ss_plot, cores_s, axes2.flat)):
    medias, _ = sortear(P, S_val, N_fixo, populacao)
    ep_s      = std_pop / np.sqrt(S_val)
    x_s       = np.linspace(media_pop - 4*ep_s, media_pop + 4*ep_s, 300)

    # Histograma
    ax.hist(medias, bins=40, density=True,
            color=cor, edgecolor='white', linewidth=0.5,
            alpha=0.85, label=f'S={S_val}')

    # Curva normal teórica para este S
    ax.plot(x_s, norm.pdf(x_s, media_pop, ep_s),
            color='#B71C1C', linewidth=2.2, linestyle='--',
            label=f'Normal (EP={ep_s:.2f})')

    # Média populacional
    ax.axvline(media_pop, color='#1A237E', linewidth=1.8,
               linestyle=':', label=f'μ = {media_pop:.1f}')

    ax.set_title(f'S = {S_val} elementos por amostra\nEP = σ/√S = {ep_s:.2f}',
                 fontsize=11, fontweight='bold')
    ax.set_xlabel('Média amostral $\\bar{x}$', fontsize=10)
    ax.set_ylabel('Densidade', fontsize=10)
    ax.legend(fontsize=8)
    ax.set_xlim(media_pop - 4*ep_s, media_pop + 4*ep_s)
    ax.spines[['top', 'right']].set_visible(False)

    # Anotação do intervalo de confiança ~95%
    ic_inf = media_pop - 2*ep_s
    ic_sup = media_pop + 2*ep_s
    ax.axvspan(ic_inf, ic_sup, alpha=0.12, color='#1A237E',
               label='IC 95% teórico')
    ax.text(0.97, 0.95,
            f'IC 95%: [{ic_inf:.1f}, {ic_sup:.1f}]\nAmplitude: {2*2*ep_s:.2f}',
            transform=ax.transAxes, fontsize=8,
            ha='right', va='top',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white',
                      edgecolor='gray', alpha=0.8))

plt.tight_layout()
plt.show()

# ══════════════════════════════════════════════════════════
# FIGURA 3 — Comparação direta: sobreposição de N e S
# ══════════════════════════════════════════════════════════
fig3, (ax_n, ax_s) = plt.subplots(1, 2, figsize=(16, 6))
fig3.suptitle('Comparação Direta — Efeito de N e S sobrepostos',
              fontsize=13, fontweight='bold')

# ── Painel esquerdo: sobreposição de diferentes N ──
Ns_comp  = [10, 50, 200, 1000, 5000]
cores_np = plt.cm.Reds(np.linspace(0.3, 0.9, len(Ns_comp)))
ep_fixo  = std_pop / np.sqrt(S_fixo)
x_fixo   = np.linspace(media_pop - 4*ep_fixo, media_pop + 4*ep_fixo, 300)

for N_val, cor in zip(Ns_comp, cores_np):
    medias_c, _ = sortear(P, S_fixo, N_val, populacao)
    ax_n.hist(medias_c, bins=30, density=True, alpha=0.45,
              color=cor, edgecolor='none', label=f'N={N_val}')

ax_n.plot(x_fixo, norm.pdf(x_fixo, media_pop, ep_fixo),
          'k--', linewidth=2.2, label='Normal teórica')
ax_n.axvline(media_pop, color='blue', linewidth=1.5,
             linestyle=':', label=f'μ = {media_pop:.1f}')
ax_n.set_title(f'Efeito de N (S fixo = {S_fixo})\n'
               f'Mais N → histograma mais suave e próximo da normal',
               fontweight='bold', fontsize=11)
ax_n.set_xlabel('Média amostral $\\bar{x}$', fontsize=11)
ax_n.set_ylabel('Densidade', fontsize=11)
ax_n.legend(fontsize=9)
ax_n.spines[['top', 'right']].set_visible(False)

# ── Painel direito: sobreposição de diferentes S ──
Ss_comp  = [2, 5, 15, 30, 50]
cores_sp = plt.cm.Greens(np.linspace(0.3, 0.9, len(Ss_comp)))

for S_val, cor in zip(Ss_comp, cores_sp):
    medias_c, _ = sortear(P, S_val, N_fixo, populacao)
    ep_s = std_pop / np.sqrt(S_val)
    ax_s.hist(medias_c, bins=40, density=True, alpha=0.45,
              color=cor, edgecolor='none',
              label=f'S={S_val} (EP={ep_s:.1f})')

ax_s.axvline(media_pop, color='blue', linewidth=1.5,
             linestyle=':', label=f'μ = {media_pop:.1f}')
ax_s.set_title(f'Efeito de S (N fixo = {N_fixo})\n'
               f'Mais S → distribuição mais estreita e precisa',
               fontweight='bold', fontsize=11)
ax_s.set_xlabel('Média amostral $\\bar{x}$', fontsize=11)
ax_s.set_ylabel('Densidade', fontsize=11)
ax_s.legend(fontsize=9)
ax_s.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()

# ══════════════════════════════════════════════════════════
# FIGURA 4 — Boxplot: S vs dispersão das médias
# ══════════════════════════════════════════════════════════
fig4, ax_bp = plt.subplots(figsize=(14, 6))

Ss_bp    = [2, 5, 10, 15, 20, 30, 50]
dados_bp = []
eps_bp   = []

for S_val in Ss_bp:
    medias_bp, _ = sortear(P, S_val, N_fixo, populacao)
    dados_bp.append(medias_bp)
    eps_bp.append(std_pop / np.sqrt(S_val))

bp = ax_bp.boxplot(dados_bp,
                   patch_artist=True,
                   labels=[f'S={s}' for s in Ss_bp],
                   medianprops=dict(color='red', linewidth=2),
                   whiskerprops=dict(linewidth=1.3),
                   capprops=dict(linewidth=1.3),
                   flierprops=dict(marker='o', markersize=3,
                                   alpha=0.4, color='#555'))

verdes = plt.cm.Greens(np.linspace(0.35, 0.85, len(Ss_bp)))
for patch, cor in zip(bp['boxes'], verdes):
    patch.set_facecolor(cor)
    patch.set_alpha(0.75)

ax_bp.axhline(media_pop, color='red', linestyle='--',
              linewidth=1.8, label=f'μ = {media_pop:.1f}')

# Anotar EP teórico em cada box
for i, (S_val, ep_s) in enumerate(zip(Ss_bp, eps_bp), 1):
    ax_bp.text(i, media_pop + ep_s + 1.5, f'EP={ep_s:.1f}',
               ha='center', fontsize=8, color='#1A237E', fontweight='bold')

ax_bp.set_title(f'Boxplot — Efeito de S na dispersão das médias amostrais (N={N_fixo})\n'
                f'Quanto maior S, mais estreita a caixa → maior precisão',
                fontweight='bold', fontsize=12)
ax_bp.set_xlabel('Tamanho da amostra (S)', fontsize=11)
ax_bp.set_ylabel('Média amostral $\\bar{x}$', fontsize=11)
ax_bp.legend(fontsize=10)
ax_bp.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
````

# Exercício 2 - Resolução sem o aparato Teórico


Traduzindo linha por linha:

Para cada um dos 10.000 experimentos:
    1. Sortear 25 números de uma Normal(186, 12)
       → simula medir 25 adultos aleatórios
    2. Calcular a média desses 25 números
       → simula obter x̄ de cada grupo
    3. Guardar essa média


### Exemplo concreto de 3 experimentos
```
Experimento 1: [182, 191, 174, 200, 188, ...]  → média = 185.3
Experimento 2: [179, 195, 183, 201, 172, ...]  → média = 187.1
Experimento 3: [190, 177, 186, 169, 194, ...]  → média = 184.8
...
Experimento 10000: [...]                        → média = 186.6
```

Ao final: temos 10.000 médias, uma por experimento




Ao plotar essas 10.000 médias, vemos:

- Uma **curva em sino** centrada em 186 — as médias oscilam em torno do valor verdadeiro
- A **largura** da curva é controlada pelo Erro Padrão $EP = s/\sqrt{n} = 2{,}4$
- Os **percentis 2,5% e 97,5%** dessa curva definem o IC empiricamente

Percentil 2,5%  →  181,05  ← 2,5% das médias ficaram abaixo disso
Percentil 97,5% →  190,95  ← 2,5% das médias ficaram acima disso
                             95% das médias ficaram DENTRO




### O que a proporção acumulada significa
```
Após   10 experimentos: 9 acertaram  → 90%  (ainda oscilando)
Após  100 experimentos: 94 acertaram → 94%  (se aproximando)
Após  500 experimentos: 476 acertaram → 95,2% (estabilizando)
Após 5000 experimentos: 4751 acertaram → 95,0% (convergiu)
```

Isso demonstra empiricamente que "95% de confiança" não é um número arbitrário — é a **frequência real** com que o método acerta quando repetido muitas vezes.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)

# ══════════════════════════════════════════════════════════
# Dados do problema
# ══════════════════════════════════════════════════════════
n     = 25
x_bar = 186
s     = 12

# ══════════════════════════════════════════════════════════
# Simulação direta: gerar 10.000 amostras de tamanho 25
# e ver quantas médias caem dentro do IC teórico
# ══════════════════════════════════════════════════════════
N     = 10_000
medias = np.array([
    np.random.normal(loc=x_bar, scale=s, size=n).mean()
    for _ in range(N)
])

# IC teórico pela fórmula
ep        = s / np.sqrt(n)
t_crit    = stats.t.ppf(0.975, df=n - 1)
ic_inf    = x_bar - t_crit * ep
ic_sup    = x_bar + t_crit * ep

# Percentis empíricos das médias simuladas
p025 = np.percentile(medias, 2.5)
p975 = np.percentile(medias, 97.5)

# ══════════════════════════════════════════════════════════
# Resultados
# ══════════════════════════════════════════════════════════
print("=" * 50)
print("  SIMULAÇÃO SIMPLES — IC DO COLESTEROL")
print("=" * 50)
print(f"  Média das médias simuladas: {medias.mean():.2f}")
print(f"  Desvio padrão das médias:   {medias.std():.4f}")
print(f"  Erro padrão teórico EP:     {ep:.4f}")
print("-" * 50)
print(f"  IC fórmula:     [{ic_inf:.2f} ; {ic_sup:.2f}]")
print(f"  IC simulação:   [{p025:.2f} ; {p975:.2f}]")
print("=" * 50)

# ══════════════════════════════════════════════════════════
# GRÁFICOS
# ══════════════════════════════════════════════════════════
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
fig.suptitle(
    f'Simulação Simples — IC 95% do Colesterol\n'
    f'n={n}  x̄={x_bar}  s={s}  →  '
    f'IC = [{ic_inf:.2f} ; {ic_sup:.2f}]',
    fontsize=13, fontweight='bold')

c_azul  = '#1565C0'
c_verm  = '#C62828'
c_verde = '#2E7D32'

# ── Painel 1 — Histograma das médias simuladas ────────────
ax1 = axes[0]
ax1.hist(medias, bins=60, density=True,
         color=c_azul, alpha=0.7, edgecolor='none')

# Curva normal teórica
x_n = np.linspace(medias.min(), medias.max(), 400)
ax1.plot(x_n, stats.norm.pdf(x_n, x_bar, ep),
         color='black', linewidth=2, label='Normal teórica')

# IC simulado (percentis)
ax1.axvline(p025, color=c_verde, linewidth=2,
            linestyle='--', label=f'2,5% empírico = {p025:.2f}')
ax1.axvline(p975, color=c_verde, linewidth=2,
            linestyle='--', label=f'97,5% empírico = {p975:.2f}')

# IC fórmula
ax1.axvline(ic_inf, color=c_verm, linewidth=2,
            linestyle=':', label=f'IC fórmula inf = {ic_inf:.2f}')
ax1.axvline(ic_sup, color=c_verm, linewidth=2,
            linestyle=':', label=f'IC fórmula sup = {ic_sup:.2f}')

# Área do IC
xic = np.linspace(p025, p975, 300)
ax1.fill_between(xic, stats.norm.pdf(xic, x_bar, ep),
                 alpha=0.15, color=c_azul, label='IC 95%')

ax1.axvline(x_bar, color='black', linestyle='-',
            linewidth=1.5, label=f'x̄ = {x_bar}')

ax1.set_title(f'Histograma de {N:,} médias simuladas\n'
              f'Cada média = média de {n} adultos sorteados',
              fontweight='bold', fontsize=10)
ax1.set_xlabel('Média do colesterol (mg/dL)', fontsize=10)
ax1.set_ylabel('Densidade', fontsize=10)
ax1.legend(fontsize=7.5)
ax1.spines[['top', 'right']].set_visible(False)



# ── Painel 3 — Convergência da proporção de captura ──────
ax3 = axes[1]
N_conv    = 5_000
capturas_conv = []

for _ in range(N_conv):
    am    = np.random.normal(x_bar, s, size=n)
    m_c   = am.mean()
    ep_c  = am.std(ddof=1) / np.sqrt(n)
    li_c  = m_c - t_crit * ep_c
    ls_c  = m_c + t_crit * ep_c
    capturas_conv.append(li_c <= x_bar <= ls_c)

prop_acum = np.cumsum(capturas_conv) / np.arange(1, N_conv + 1) * 100

ax3.plot(range(1, N_conv + 1), prop_acum,
         color=c_azul, linewidth=1.8,
         label='Proporção acumulada de capturas')
ax3.axhline(95, color=c_verm, linestyle='--',
            linewidth=2, label='95% esperado')
ax3.fill_between(range(1, N_conv + 1),
                 93, 97, alpha=0.1, color=c_verm,
                 label='Faixa ±2% em torno de 95%')

ax3.set_title('Convergência — proporção de ICs\nque capturam μ',
              fontweight='bold', fontsize=10)
ax3.set_xlabel('Número de simulações', fontsize=10)
ax3.set_ylabel('% de ICs que capturam μ', fontsize=10)
ax3.set_ylim(85, 100)
ax3.legend(fontsize=8)
ax3.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
````

# Exercício 2 (Teórico Resolucão)

**O Ponto Central**


Quando a população é normal, as médias amostrais são normais por herança matemática, não por aproximação. 

---

## 1. Lendo o Problema

> *"Uma amostra de 25 adultos do sexo masculino mostrou que o nível de colesterol tem média 186 e desvio padrão 12. A população segue uma normal. Obter o IC de 95%."*

| O que o problema diz | Símbolo | Valor |
|---|---|---|
| Tamanho da amostra | $n$ | $25$ |
| Média amostral | $\bar{x}$ | $186$ mg/dL |
| Desvio padrão **amostral** | $s$ | $12$ mg/dL |
| Desvio padrão **populacional** | $\sigma$ | ❌ Desconhecido |
| Nível de confiança | $1 - \alpha$ | $95\%$ |

---

## 2. Primeira Decisão — Z ou t?

Antes de calcular qualquer coisa, precisamos escolher a distribuição correta.

### A lógica da padronização

Qualquer estimativa estatística pode ser padronizada pela fórmula geral:

$$\text{estatística} = \frac{\text{estimativa} - \text{parâmetro verdadeiro}}{\text{desvio padrão da estimativa}}$$

Para a média amostral $\bar{x}$, sabemos que:

$$E(\bar{x}) = \mu \qquad \text{Var}(\bar{x}) = \frac{\sigma^2}{n} \qquad \text{DP}(\bar{x}) = \frac{\sigma}{\sqrt{n}}$$

Padronizando:

$$\frac{\bar{x} - \mu}{\sigma/\sqrt{n}}$$

### O problema: $\sigma$ é desconhecido

Se $\sigma$ fosse conhecido, essa expressão seguiria exatamente $\mathcal{N}(0,1)$ e usaríamos $Z$. Mas $\sigma$ **não é conhecido** — só temos $s = 12$ da amostra.

A tentação é simplesmente substituir $\sigma$ por $s$:

$$\frac{\bar{x} - \mu}{s/\sqrt{n}}$$

Mas isso **não segue mais a Normal** — porque $s$ é aleatório e carrega incerteza extra.

### Como o t de Student é construído?

O $t$ emerge da combinação de duas distribuições independentes:

**Componente 1** — Normal padrão (numerador):
$$Z = \frac{\bar{x} - \mu}{\sigma/\sqrt{n}} \sim \mathcal{N}(0,1)$$

**Componente 2** — Qui-quadrado (denominador):
$$V = \frac{(n-1)s^2}{\sigma^2} \sim \chi^2_{(n-1)}$$

Combinando pela definição formal do $t$:

$$t = \frac{Z}{\sqrt{V/(n-1)}} = \frac{\dfrac{\bar{x}-\mu}{\sigma/\sqrt{n}}}{\dfrac{s}{\sigma}} = \frac{\bar{x} - \mu}{s/\sqrt{n}}$$

Os $\sigma$ **se cancelam** — o resultado não depende de $\sigma$, que era exatamente o que queríamos eliminar!

$$\boxed{t = \frac{\bar{x} - \mu}{s/\sqrt{n}} \sim t_{(n-1)}}$$

### Resumo da decisão

```
σ populacional conhecido?
          │
     ┌────┴────┐
    SIM        NÃO  ← nosso caso
     │          │
     ▼          ▼
    Z = x̄-μ   t = x̄-μ
        ─────      ─────
        σ/√n       s/√n

  ~ N(0,1)     ~ t(n-1)
  Caudas       Caudas mais
  finas        pesadas
```

---

## 3. Graus de Liberdade — Por que $n-1$?

$$gl = n - 1 = 25 - 1 = 24$$

Quando calculamos $s$, já usamos $\bar{x}$ para estimar a média. Isso **consome 1 grau de liberdade** — os $n$ valores não são mais totalmente livres, pois precisam satisfazer:

$$\sum_{i=1}^{n}(x_i - \bar{x}) = 0$$

Se você conhece $n-1$ valores e conhece $\bar{x}$, o último valor fica **automaticamente determinado**. Por isso sobram apenas $n-1$ graus livres.

**Efeito prático:** quanto menor o $gl$, mais largas são as caudas do $t$ — o intervalo fica mais conservador para compensar a maior incerteza.

---

## 4. Nível de Significância e as Caudas

$$\alpha = 1 - 0{,}95 = 0{,}05$$

Como queremos capturar $\mu$ de **ambos os lados** de $\bar{x}$, o teste é **bicaudal** — dividimos $\alpha$ igualmente:

$$\frac{\alpha}{2} = \frac{0{,}05}{2} = 0{,}025 \quad \text{em cada cauda}$$

```
        2,5%            95%             2,5%
   ──────────┬───────────────────────┬──────────
             │                       │
          -t crítico               +t crítico
```

Buscamos $t$ tal que $P(-t \leq t_{24} \leq +t) = 0{,}95$.

---

## 5. Por que as Caudas do t são mais Pesadas que Z?

A função densidade do $t_{(k)}$ decai **polinomialmente**:

$$f(t) \propto \left(1 + \frac{t^2}{k}\right)^{-\frac{k+1}{2}}$$

Enquanto a Normal decai **exponencialmente**:

$$f(z) \propto e^{-z^2/2}$$

O decaimento polinomial é muito mais lento — valores extremos são mais prováveis no $t$ do que na Normal. Isso se traduz em **valores críticos maiores** e **intervalos mais largos**.

Conforme $gl \to \infty$, o $t$ converge para $Z$:

| $gl$ | $t_{0{,}025}$ | Diferença para $Z = 1{,}96$ |
|---|---|---|
| 5 | $2{,}5706$ | $+0{,}61$ |
| 10 | $2{,}2281$ | $+0{,}27$ |
| **24** | **$2{,}0639$** | **$+0{,}10$** ← nosso caso |
| 30 | $2{,}0423$ | $+0{,}08$ |
| 60 | $2{,}0003$ | $+0{,}04$ |
| $\infty$ | $1{,}9600$ | $0{,}00$ |

---

## 6. Cálculo do Erro Padrão

$$EP = \frac{s}{\sqrt{n}} = \frac{12}{\sqrt{25}} = \frac{12}{5} = 2{,}4 \text{ mg/dL}$$

O EP representa a **dispersão esperada das médias amostrais** em torno de $\mu$. Em outras palavras: se repetíssemos esse experimento muitas vezes, as médias oscilariam em torno de $\mu$ com desvio padrão de $2{,}4$ mg/dL.

---

## 7. Valor Crítico $t$

Para $gl = 24$ e $\alpha/2 = 0{,}025$:

$$t_{0{,}025;\ 24} = 2{,}0639$$

Isso significa: 95% da área da distribuição $t_{(24)}$ está entre $-2{,}0639$ e $+2{,}0639$.

---

## 8. Margem de Erro

$$ME = t_{\alpha/2;\ n-1} \times EP = 2{,}0639 \times 2{,}4 = 4{,}9533 \text{ mg/dL}$$

A margem de erro é o **raio** do intervalo — o quanto nos afastamos de $\bar{x}$ para cada lado.

---

## 9. Construindo o Intervalo

$$IC_{95\%} = \bar{x} \pm ME = 186 \pm 4{,}9533$$

| Limite | Cálculo | Resultado |
|---|---|---|
| Inferior | $186 - 4{,}9533$ | $181{,}05$ mg/dL |
| Superior | $186 + 4{,}9533$ | $190{,}95$ mg/dL |
| Amplitude | $2 \times 4{,}9533$ | $9{,}91$ mg/dL |

$$\boxed{IC_{95\%} = [181{,}05\ ;\ 190{,}95] \text{ mg/dL}}$$

---


## 12. O que cada gráfico mostra

**Painel 1 — Z vs t:** a distribuição $t$ tem caudas mais pesadas que a Normal — região laranja mostra o excesso. Os valores críticos $1{,}96$ e $2{,}0639$ mostram que o $t$ é mais conservador.

**Painel 2 — Distribuição t com regiões:** a área central azul (95%) é delimitada pelos valores críticos $\pm 2{,}0639$. As caudas vermelhas representam os 5% de chance de erro.

**Painel 3 — IC na escala original:** o intervalo $[181{,}05\ ;\ 190{,}95]$ centrado em $\bar{x} = 186$, com a amplitude total de $9{,}91$ mg/dL e margem de erro de $4{,}95$ mg/dL destacadas.

**Painel 4 — Efeito dos graus de liberdade:** conforme $gl$ cresce, o $t$ crítico diminui e o IC encolhe — convergindo para o IC baseado em $Z$ quando $gl \to \infty$.

---

## 13. Interpretação Final

$$\boxed{IC_{95\%} = [181{,}05\ ;\ 190{,}95] \text{ mg/dL}}$$

> **"Temos 95% de confiança de que o verdadeiro nível médio de colesterol da população de adultos masculinos nessa cidade está entre $181{,}05$ e $190{,}95$ mg/dL."**

| ✅ Correto | ❌ Errado |
|---|---|
| 95% dos **intervalos** assim construídos conterão $\mu$ | 95% dos **indivíduos** têm colesterol nesse intervalo |
| $\mu$ é fixo — o **intervalo** é que é aleatório | Há 95% de probabilidade de $\mu$ estar nesse intervalo |

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# ══════════════════════════════════════════════════════════
# Dados do problema
# ══════════════════════════════════════════════════════════
n       = 25       # tamanho da amostra
x_bar   = 186      # média amostral
s       = 12       # desvio padrão amostral
confi   = 0.95     # nível de confiança
alpha   = 1 - confi  # nível de significância = 0.05
gl      = n - 1    # graus de liberdade = 24

# ══════════════════════════════════════════════════════════
# Cálculo do Intervalo de Confiança
# ══════════════════════════════════════════════════════════
ep        = s / np.sqrt(n)               # Erro Padrão = 12/√25 = 2.4
t_critico = stats.t.ppf(1 - alpha/2, gl) # t crítico bicaudal (α/2 = 0.025)
margem    = t_critico * ep               # Margem de erro
ic_inf    = x_bar - margem               # Limite inferior
ic_sup    = x_bar + margem               # Limite superior

# ══════════════════════════════════════════════════════════
# Resultados
# ══════════════════════════════════════════════════════════
print("=" * 52)
print("   INTERVALO DE CONFIANÇA — COLESTEROL")
print("=" * 52)
print(f"   Tamanho da amostra (n):      {n}")
print(f"   Média amostral (x̄):          {x_bar}")
print(f"   Desvio padrão amostral (s):  {s}")
print(f"   Nível de confiança:          {confi*100:.0f}%")
print(f"   Nível de significância (α):  {alpha:.2f}")
print(f"   Graus de liberdade:          {gl}")
print("-" * 52)
print(f"   Erro Padrão (s/√n):          {ep:.4f}")
print(f"   t crítico (t₀.₀₂₅, ₂₄):     {t_critico:.4f}")
print(f"   Margem de erro:              {margem:.4f}")
print("-" * 52)
print(f"   Limite inferior (IC):        {ic_inf:.4f}")
print(f"   Limite superior (IC):        {ic_sup:.4f}")
print("=" * 52)
print(f"\n   IC 95% = [{ic_inf:.2f} ;  {ic_sup:.2f}]")
print(f"\n   Interpretação:")
print(f"   Temos 95% de confiança de que a média")
print(f"   populacional do colesterol está entre")
print(f"   {ic_inf:.2f} e {ic_sup:.2f} mg/dL.")

# ══════════════════════════════════════════════════════════
# GRÁFICO — Distribuição t com IC destacado
# ══════════════════════════════════════════════════════════
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Intervalo de Confiança 95% — Nível de Colesterol',
             fontsize=14, fontweight='bold')

# ── Painel 1: Distribuição t com regiões destacadas ──────
x = np.linspace(-4.5, 4.5, 500)
y = stats.t.pdf(x, gl)

ax1.plot(x, y, color='#1565C0', linewidth=2.5, label=f't-Student (gl={gl})')

# Região de confiança (95%) — centro
x_conf = np.linspace(-t_critico, t_critico, 300)
ax1.fill_between(x_conf, stats.t.pdf(x_conf, gl),
                 alpha=0.3, color='#1565C0', label='95% de confiança')

# Regiões de rejeição (α/2 = 2.5% cada lado)
x_rej_e = np.linspace(-4.5, -t_critico, 100)
x_rej_d = np.linspace(t_critico, 4.5, 100)
ax1.fill_between(x_rej_e, stats.t.pdf(x_rej_e, gl),
                 alpha=0.4, color='#E53935', label='α/2 = 2.5% (cada cauda)')
ax1.fill_between(x_rej_d, stats.t.pdf(x_rej_d, gl),
                 alpha=0.4, color='#E53935')

# Linhas dos valores críticos
ax1.axvline(-t_critico, color='#E53935', linestyle='--',
            linewidth=1.8, label=f'±t crítico = ±{t_critico:.4f}')
ax1.axvline( t_critico, color='#E53935', linestyle='--', linewidth=1.8)
ax1.axvline(0, color='#2E7D32', linestyle=':', linewidth=1.5, label='Centro (t=0)')

# Anotações
ax1.text( t_critico + 0.1, 0.28, f'+{t_critico:.3f}',
          color='#E53935', fontsize=10, fontweight='bold')
ax1.text(-t_critico - 0.5, 0.28, f'-{t_critico:.3f}',
          color='#E53935', fontsize=10, fontweight='bold')
ax1.text(0, 0.22, '95%', ha='center', fontsize=13,
         color='#1565C0', fontweight='bold')
ax1.text(-3.5, 0.06, '2.5%', ha='center', fontsize=10, color='#E53935')
ax1.text( 3.5, 0.06, '2.5%', ha='center', fontsize=10, color='#E53935')

ax1.set_title(f'Distribuição t-Student (gl = {gl})\nRegiões de confiança e rejeição',
              fontweight='bold', fontsize=11)
ax1.set_xlabel('Valor de t', fontsize=11)
ax1.set_ylabel('Densidade de probabilidade', fontsize=11)
ax1.legend(fontsize=9)
ax1.spines[['top', 'right']].set_visible(False)

# ── Painel 2: IC na escala original (mg/dL) ──────────────
ax2.axvline(x_bar, color='#1565C0', linewidth=2.5,
            label=f'Média amostral x̄ = {x_bar}')
ax2.axvline(ic_inf, color='#E53935', linewidth=2,
            linestyle='--', label=f'Limite inferior = {ic_inf:.2f}')
ax2.axvline(ic_sup, color='#2E7D32', linewidth=2,
            linestyle='--', label=f'Limite superior = {ic_sup:.2f}')

# Faixa do IC
ax2.axvspan(ic_inf, ic_sup, alpha=0.15, color='#1565C0',
            label=f'IC 95% = [{ic_inf:.2f} ; {ic_sup:.2f}]')

# Seta mostrando a amplitude
ax2.annotate('', xy=(ic_sup, 0.5), xytext=(ic_inf, 0.5),
             arrowprops=dict(arrowstyle='<->', color='black',
                             lw=1.8))
ax2.text(x_bar, 0.53, f'Amplitude = {ic_sup - ic_inf:.2f} mg/dL',
         ha='center', fontsize=10, fontweight='bold')

# Margem de erro
ax2.annotate('', xy=(x_bar + margem, 0.35), xytext=(x_bar, 0.35),
             arrowprops=dict(arrowstyle='<->', color='#E53935', lw=1.5))
ax2.text(x_bar + margem/2, 0.38, f'Margem\n={margem:.2f}',
         ha='center', fontsize=9, color='#E53935')

# Distribuição normal da média amostral
x2 = np.linspace(ic_inf - 15, ic_sup + 15, 400)
y2 = stats.t.pdf((x2 - x_bar) / ep, gl) / ep
ax2.plot(x2, y2 / max(y2) * 0.8, color='#1565C0',
         linewidth=2, alpha=0.5, linestyle='-')

ax2.set_xlim(ic_inf - 15, ic_sup + 15)
ax2.set_ylim(0, 0.75)
ax2.set_title(f'IC 95% na escala do colesterol (mg/dL)\n'
              f'x̄ ± t × EP = {x_bar} ± {t_critico:.4f} × {ep:.1f}',
              fontweight='bold', fontsize=11)
ax2.set_xlabel('Nível de colesterol (mg/dL)', fontsize=11)
ax2.set_ylabel('', fontsize=11)
ax2.legend(fontsize=9, loc='upper right')
ax2.spines[['top', 'right']].set_visible(False)

# Anotações dos limites
ax2.text(ic_inf, 0.68, f'{ic_inf:.2f}', ha='center',
         fontsize=10, color='#E53935', fontweight='bold')
ax2.text(ic_sup, 0.68, f'{ic_sup:.2f}', ha='center',
         fontsize=10, color='#2E7D32', fontweight='bold')
ax2.text(x_bar,  0.68, f'{x_bar}',      ha='center',
         fontsize=10, color='#1565C0', fontweight='bold')

plt.tight_layout()
plt.show()
````



## Erros Tipo I, Tipo II e o p-valor

---

## 1. A Situação

Imagine que um laboratório afirma que um medicamento **não altera** o colesterol. Você coleta uma amostra e quer testar essa afirmação.

O teste de hipótese funciona como um **tribunal**:

```
H₀ (Hipótese Nula)        → "O medicamento não tem efeito"
                             → O réu é INOCENTE até prova em contrário

H₁ (Hipótese Alternativa) → "O medicamento tem efeito"
                             → O réu é CULPADO
```

Assim como num tribunal, podemos **errar a sentença**:

---

## 2. Os Dois Tipos de Erro

| | H₀ verdadeira (inocente) | H₀ falsa (culpado) |
|---|---|---|
| **Rejeitar H₀** | ❌ Erro Tipo I ($\alpha$) | ✅ Correto |
| **Não rejeitar H₀** | ✅ Correto | ❌ Erro Tipo II ($\beta$) |

**Erro Tipo I ($\alpha$):** condenar um inocente — rejeitar H₀ quando ela é verdadeira
**Erro Tipo II ($\beta$):** absolver um culpado — não rejeitar H₀ quando ela é falsa

---

## 3. O p-valor

> O p-valor é a probabilidade de obter um resultado **tão extremo quanto o observado**, assumindo que H₀ é verdadeira.

- **p-valor pequeno** → o resultado é improvável se H₀ for verdadeira → evidência contra H₀
- **p-valor grande** → o resultado é plausível mesmo se H₀ for verdadeira → sem evidência contra H₀


### O que cada figura mostra

**Figura 1 — Os 3 cenários:**
mostra a distribuição $t$ com as regiões críticas em vermelho. A área laranja é o **p-valor** — quanto mais o $t$ observado se afasta do centro, menor o p-valor e mais forte a evidência contra H₀.

**Figura 2 — Erro Tipo I e II:**
as duas distribuições são a de H₀ e a de H₁. O **Erro Tipo I** ($\alpha$) é a cauda vermelha de H₀ além da região crítica. O **Erro Tipo II** ($\beta$) é a área laranja de H₁ dentro da região de não rejeição. O **Poder** ($1 - \beta$) é a área verde de H₁ além da região crítica — a capacidade do teste de detectar o efeito real.

**Figura 3 — p-valor variando:**
mostra como o p-valor cai conforme $\bar{x}$ se afasta de 186. A linha preta ($\alpha = 0{,}05$) é o limite de decisão — abaixo dela rejeitamos H₀. Os três cenários do exercício são marcados com pontos maiores.

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy import stats

np.random.seed(42)

# ══════════════════════════════════════════════════════════
# Contexto do problema
# ══════════════════════════════════════════════════════════
# H₀: μ = 186  (medicamento não altera o colesterol)
# H₁: μ ≠ 186  (medicamento altera o colesterol)
# Dados da amostra coletada após tomar o medicamento:
mu_0   = 186     # média sob H₀
n      = 25      # tamanho da amostra
alpha  = 0.05    # nível de significância
gl     = n - 1   # graus de liberdade

# ══════════════════════════════════════════════════════════
# 3 Cenários para ilustrar os conceitos
# ══════════════════════════════════════════════════════════
cenarios = {
    'Rejeita H₀\n(p < α)\nEfeito real':
        {'x_bar': 181.0, 's': 12, 'mu_real': 181},
    'Não rejeita H₀\n(p > α)\nSem efeito':
        {'x_bar': 184.2, 's': 12, 'mu_real': 186},
    'Limítrofe\n(p ≈ α)\nZona cinzenta':
        {'x_bar': 182.0, 's': 12, 'mu_real': 181},
}

def calcular_teste(x_bar, s, n, mu_0):
    ep      = s / np.sqrt(n)
    t_stat  = (x_bar - mu_0) / ep
    p_valor = 2 * stats.t.sf(abs(t_stat), df=n - 1)
    t_crit  = stats.t.ppf(1 - alpha / 2, df=n - 1)
    rejeita = p_valor < alpha
    return {'ep': ep, 't_stat': t_stat,
            'p_valor': p_valor, 't_crit': t_crit,
            'rejeita': rejeita}

# ══════════════════════════════════════════════════════════
# Resultados no console
# ══════════════════════════════════════════════════════════
print("=" * 65)
print("  TESTE DE HIPÓTESE — EFEITO DO MEDICAMENTO NO COLESTEROL")
print(f"  H₀: μ = {mu_0}  |  H₁: μ ≠ {mu_0}  |  α = {alpha}")
print("=" * 65)
for nome, c in cenarios.items():
    r = calcular_teste(c['x_bar'], c['s'], n, mu_0)
    print(f"\n  Cenário: {nome.replace(chr(10), ' ')}")
    print(f"  x̄ = {c['x_bar']}  s = {c['s']}")
    print(f"  t calculado = {r['t_stat']:.4f}")
    print(f"  t crítico   = ±{r['t_crit']:.4f}")
    print(f"  p-valor     = {r['p_valor']:.4f}")
    print(f"  Decisão     → {'REJEITA H₀ ✓' if r['rejeita'] else 'NÃO REJEITA H₀'}")
print("=" * 65)

# ══════════════════════════════════════════════════════════
# FIGURA 1 — Os 3 cenários com p-valor ilustrado
# ══════════════════════════════════════════════════════════
fig1, axes = plt.subplots(1, 3, figsize=(18, 6))
fig1.suptitle('p-valor nos 3 Cenários — H₀: μ = 186',
              fontsize=13, fontweight='bold')

x_t = np.linspace(-5, 5, 600)
y_t = stats.t.pdf(x_t, df=gl)

for ax, (nome, c) in zip(axes, cenarios.items()):
    r = calcular_teste(c['x_bar'], c['s'], n, mu_0)
    tc   = r['t_crit']
    tobs = r['t_stat']
    pval = r['p_valor']

    # Curva t
    ax.plot(x_t, y_t, color='#1565C0', linewidth=2.5)

    # Regiões críticas (α/2 em cada cauda)
    x_rc_d = np.linspace(tc, 5, 200)
    x_rc_e = np.linspace(-5, -tc, 200)
    ax.fill_between(x_rc_d, stats.t.pdf(x_rc_d, gl),
                    alpha=0.4, color='#C62828',
                    label=f'Região crítica (α={alpha})')
    ax.fill_between(x_rc_e, stats.t.pdf(x_rc_e, gl),
                    alpha=0.4, color='#C62828')

    # p-valor (área além do t observado)
    x_pv_d = np.linspace(abs(tobs), 5, 200)
    x_pv_e = np.linspace(-5, -abs(tobs), 200)
    ax.fill_between(x_pv_d, stats.t.pdf(x_pv_d, gl),
                    alpha=0.6, color='#E65100',
                    label=f'p-valor = {pval:.4f}')
    ax.fill_between(x_pv_e, stats.t.pdf(x_pv_e, gl),
                    alpha=0.6, color='#E65100')

    # t observado
    cor_tobs = '#C62828' if r['rejeita'] else '#2E7D32'
    ax.axvline(tobs, color=cor_tobs, linewidth=2.5,
               linestyle='--',
               label=f't obs = {tobs:.3f}')
    ax.axvline(-tc, color='#C62828', linewidth=1.5,
               linestyle=':', alpha=0.7,
               label=f't crítico = ±{tc:.3f}')
    ax.axvline( tc, color='#C62828', linewidth=1.5,
               linestyle=':', alpha=0.7)

    # Decisão
    decisao   = 'REJEITA H₀' if r['rejeita'] else 'NÃO REJEITA H₀'
    cor_dec   = '#C62828'     if r['rejeita'] else '#2E7D32'
    ax.text(0, stats.t.pdf(0, gl) * 0.4, decisao,
            ha='center', fontsize=11, fontweight='bold',
            color=cor_dec,
            bbox=dict(boxstyle='round,pad=0.3',
                      facecolor='white',
                      edgecolor=cor_dec, alpha=0.9))

    ax.set_title(f'{nome}\nx̄={c["x_bar"]}  p={pval:.4f}',
                 fontweight='bold', fontsize=10)
    ax.set_xlabel('Estatística t', fontsize=10)
    ax.set_ylabel('Densidade', fontsize=10)
    ax.legend(fontsize=7.5)
    ax.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()

# ══════════════════════════════════════════════════════════
# FIGURA 2 — Erro Tipo I e Tipo II
# ══════════════════════════════════════════════════════════
fig2, axes2 = plt.subplots(1, 2, figsize=(16, 7))
fig2.suptitle('Erro Tipo I (α) e Erro Tipo II (β)',
              fontsize=13, fontweight='bold')

# Parâmetros para ilustrar os dois erros
mu_H0    = 186      # média sob H₀
mu_H1    = 181      # média sob H₁ (efeito real)
sigma    = 12
ep_plot  = sigma / np.sqrt(n)
tc       = stats.t.ppf(1 - alpha / 2, df=gl) * ep_plot
x_plot   = np.linspace(mu_H1 - 5*ep_plot,
                        mu_H0 + 5*ep_plot, 800)

# Distribuições de x̄ sob H₀ e sob H₁
y_H0 = stats.norm.pdf(x_plot, mu_H0, ep_plot)
y_H1 = stats.norm.pdf(x_plot, mu_H1, ep_plot)

lim_inf = mu_H0 - tc
lim_sup = mu_H0 + tc

# ── Painel esquerdo — Erro Tipo I ─────────────────────────
ax_e1 = axes2[0]
ax_e1.plot(x_plot, y_H0, color='#1565C0',
           linewidth=2.5, label='Distribuição sob H₀')
ax_e1.plot(x_plot, y_H1, color='#2E7D32',
           linewidth=2, linestyle='--',
           alpha=0.4, label='Distribuição sob H₁')

# Erro Tipo I: caudas de H₀ além das regiões críticas
x_err1_d = x_plot[x_plot >= lim_sup]
x_err1_e = x_plot[x_plot <= lim_inf]
ax_e1.fill_between(x_err1_d,
                   stats.norm.pdf(x_err1_d, mu_H0, ep_plot),
                   alpha=0.7, color='#C62828',
                   label=f'Erro Tipo I (α = {alpha})')
ax_e1.fill_between(x_err1_e,
                   stats.norm.pdf(x_err1_e, mu_H0, ep_plot),
                   alpha=0.7, color='#C62828')

ax_e1.axvline(lim_inf, color='#C62828', linewidth=2,
              linestyle='--', label=f'Limites críticos\n[{lim_inf:.1f}; {lim_sup:.1f}]')
ax_e1.axvline(lim_sup, color='#C62828', linewidth=2,
              linestyle='--')
ax_e1.axvline(mu_H0, color='#1565C0', linewidth=1.5,
              linestyle=':', label=f'μ₀ = {mu_H0}')

ax_e1.text(mu_H0, max(y_H0) * 0.5,
           f'H₀ verdadeira\nmas rejeitamos\n→ Erro Tipo I\nP = α = {alpha}',
           ha='center', fontsize=9,
           bbox=dict(boxstyle='round,pad=0.3',
                     facecolor='#FFEBEE',
                     edgecolor='#C62828'))

ax_e1.set_title('Erro Tipo I (α)\n'
                '"Condenar um inocente"\n'
                'Rejeitar H₀ quando ela é VERDADEIRA',
                fontweight='bold', fontsize=10)
ax_e1.set_xlabel('Média amostral x̄ (mg/dL)', fontsize=10)
ax_e1.set_ylabel('Densidade', fontsize=10)
ax_e1.legend(fontsize=8, loc='upper left')
ax_e1.spines[['top', 'right']].set_visible(False)

# ── Painel direito — Erro Tipo II ─────────────────────────
ax_e2 = axes2[1]
ax_e2.plot(x_plot, y_H0, color='#1565C0',
           linewidth=2, linestyle='--',
           alpha=0.5, label=f'Distribuição sob H₀ (μ={mu_H0})')
ax_e2.plot(x_plot, y_H1, color='#2E7D32',
           linewidth=2.5, label=f'Distribuição sob H₁ (μ={mu_H1})')

# Erro Tipo II: área de H₁ dentro da região de não rejeição
x_beta = x_plot[(x_plot >= lim_inf) & (x_plot <= lim_sup)]
ax_e2.fill_between(x_beta,
                   stats.norm.pdf(x_beta, mu_H1, ep_plot),
                   alpha=0.6, color='#E65100',
                   label='Erro Tipo II (β)')

# Poder do teste (1 - β)
beta  = stats.norm.cdf(lim_sup, mu_H1, ep_plot) - \
        stats.norm.cdf(lim_inf, mu_H1, ep_plot)
poder = 1 - beta

x_pod_e = x_plot[x_plot <= lim_inf]
x_pod_d = x_plot[x_plot >= lim_sup]
ax_e2.fill_between(x_pod_e,
                   stats.norm.pdf(x_pod_e, mu_H1, ep_plot),
                   alpha=0.35, color='#2E7D32',
                   label=f'Poder = 1-β = {poder:.2f}')
ax_e2.fill_between(x_pod_d,
                   stats.norm.pdf(x_pod_d, mu_H1, ep_plot),
                   alpha=0.35, color='#2E7D32')

ax_e2.axvline(lim_inf, color='#C62828', linewidth=2,
              linestyle='--',
              label=f'Limites críticos\n[{lim_inf:.1f}; {lim_sup:.1f}]')
ax_e2.axvline(lim_sup, color='#C62828', linewidth=2,
              linestyle='--')
ax_e2.axvline(mu_H1, color='#2E7D32', linewidth=1.5,
              linestyle=':', label=f'μ₁ = {mu_H1}')

ax_e2.text(mu_H1 + 1.5, max(y_H1) * 0.45,
           f'H₀ falsa\nmas não rejeitamos\n→ Erro Tipo II\nβ = {beta:.2f}',
           ha='center', fontsize=9,
           bbox=dict(boxstyle='round,pad=0.3',
                     facecolor='#FFF3E0',
                     edgecolor='#E65100'))

ax_e2.set_title(f'Erro Tipo II (β)\n'
                f'"Absolver um culpado"\n'
                f'Não rejeitar H₀ quando ela é FALSA\n'
                f'β = {beta:.2f}  |  Poder = {poder:.2f}',
                fontweight='bold', fontsize=10)
ax_e2.set_xlabel('Média amostral x̄ (mg/dL)', fontsize=10)
ax_e2.set_ylabel('Densidade', fontsize=10)
ax_e2.legend(fontsize=8, loc='upper right')
ax_e2.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()

# ══════════════════════════════════════════════════════════
# FIGURA 3 — p-valor variando conforme x̄ muda
# ══════════════════════════════════════════════════════════
fig3, (ax_p1, ax_p2) = plt.subplots(1, 2, figsize=(16, 6))
fig3.suptitle('Como o p-valor muda conforme a média amostral se afasta de H₀',
              fontsize=12, fontweight='bold')

x_bars  = np.linspace(178, 194, 300)
p_vals  = [calcular_teste(xb, 12, n, mu_0)['p_valor']
           for xb in x_bars]
t_stats = [calcular_teste(xb, 12, n, mu_0)['t_stat']
           for xb in x_bars]

# Painel esquerdo — p-valor vs x̄
cores_p = ['#C62828' if p < alpha else '#2E7D32'
           for p in p_vals]
ax_p1.scatter(x_bars, p_vals, c=cores_p, s=15, zorder=3)
ax_p1.axhline(alpha, color='black', linewidth=2,
              linestyle='--', label=f'α = {alpha}')
ax_p1.axvline(mu_0, color='#1565C0', linewidth=1.5,
              linestyle=':', label=f'H₀: μ = {mu_0}')

ax_p1.fill_between(x_bars, 0, alpha,
                   where=[p < alpha for p in p_vals],
                   alpha=0.15, color='#C62828',
                   label='Rejeita H₀ (p < α)')
ax_p1.fill_between(x_bars, alpha, 1,
                   where=[p >= alpha for p in p_vals],
                   alpha=0.1, color='#2E7D32',
                   label='Não rejeita H₀ (p ≥ α)')

# Marcar os 3 cenários
for nome, c in cenarios.items():
    r = calcular_teste(c['x_bar'], c['s'], n, mu_0)
    ax_p1.scatter(c['x_bar'], r['p_valor'],
                  s=150, zorder=5,
                  color='#C62828' if r['rejeita'] else '#2E7D32',
                  edgecolors='black', linewidth=1.5)
    ax_p1.annotate(f"x̄={c['x_bar']}\np={r['p_valor']:.3f}",
                   (c['x_bar'], r['p_valor']),
                   textcoords='offset points',
                   xytext=(10, 5), fontsize=8)

ax_p1.set_title('p-valor em função de x̄\n'
                'Quanto mais longe de H₀, menor o p-valor',
                fontweight='bold', fontsize=10)
ax_p1.set_xlabel('Média amostral x̄ (mg/dL)', fontsize=10)
ax_p1.set_ylabel('p-valor', fontsize=10)
ax_p1.legend(fontsize=8)
ax_p1.spines[['top', 'right']].set_visible(False)

# Painel direito — Tabela resumo dos conceitos
ax_p2.axis('off')
tabela = [
    ['Conceito', 'Definição', 'Analogia'],
    ['H₀', 'Hipótese que\nqueremos testar', 'Réu é inocente'],
    ['H₁', 'Hipótese\nalternativa', 'Réu é culpado'],
    ['α', 'P(Erro Tipo I)\n= P(rejeitar H₀|H₀ verdadeira)', 'P(condenar inocente)'],
    ['β', 'P(Erro Tipo II)\n= P(não rejeitar H₀|H₀ falsa)', 'P(absolver culpado)'],
    ['1 - β', 'Poder do teste\n= P(rejeitar H₀|H₀ falsa)', 'P(condenar culpado)'],
    ['p-valor', 'P(resultado tão extremo\n| H₀ verdadeira)', 'Força da evidência\ncontra H₀'],
    ['Decisão', 'p < α → rejeita H₀\np ≥ α → não rejeita', 'Culpado / Inocente'],
]

tb = ax_p2.table(
    cellText=tabela[1:],
    colLabels=tabela[0],
    loc='center',
    cellLoc='center'
)
tb.auto_set_font_size(False)
tb.set_fontsize(9)
tb.scale(1.2, 2.8)

# Colorir cabeçalho
for j in range(3):
    tb[(0, j)].set_facecolor('#1565C0')
    tb[(0, j)].set_text_props(color='white', fontweight='bold')

# Colorir linhas alternadas
cores_tab = ['#E3F2FD', '#FFFFFF']
for i in range(1, len(tabela)):
    for j in range(3):
        tb[(i, j)].set_facecolor(cores_tab[i % 2])

ax_p2.set_title('Resumo dos Conceitos',
                fontweight='bold', fontsize=11, pad=20)

plt.tight_layout()
plt.show()
````

## Teste A/B em Software — Teste de Hipótese na Prática

---

## 1. O Cenário

Uma empresa de e-commerce quer saber se uma **nova versão do botão de compra** (versão B) converte mais usuários do que a versão atual (versão A).

```
Versão A (controle)  →  Botão cinza  "Comprar"
Versão B (tratamento) →  Botão verde  "Comprar agora!"
```

Durante 2 semanas, metade dos usuários vê A e metade vê B. Medimos a **taxa de conversão** — proporção de usuários que completaram a compra.

---

## 2. Formulando o Teste

$$H_0: p_A = p_B \quad \text{(os botões convertem igual)}$$
$$H_1: p_A \neq p_B \quad \text{(os botões convertem diferente)}$$

Aqui trabalhamos com **proporções** — não com médias. Isso muda a estatística de teste.

---

## 3. A Estatística de Teste para Proporções

$$z = \frac{\hat{p}_A - \hat{p}_B}{\sqrt{\hat{p}(1-\hat{p})\left(\frac{1}{n_A} + \frac{1}{n_B}\right)}}$$

Onde $\hat{p}$ é a proporção combinada (pooled):

$$\hat{p} = \frac{x_A + x_B}{n_A + n_B}$$



````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy import stats

np.random.seed(42)

# ══════════════════════════════════════════════════════════
# Dados do experimento A/B
# ══════════════════════════════════════════════════════════
n_A   = 2300      # usuários que viram versão A
n_B   = 2300      # usuários que viram versão B
x_A   = 230       # conversões na versão A
x_B   = 276       # conversões na versão B
alpha = 0.05      # nível de significância

# ══════════════════════════════════════════════════════════
# Cálculos
# ══════════════════════════════════════════════════════════
p_A   = x_A / n_A                         # taxa A  = 10%
p_B   = x_B / n_B                         # taxa B  = 12%
p_hat = (x_A + x_B) / (n_A + n_B)        # proporção combinada

# Erro padrão da diferença
ep    = np.sqrt(p_hat * (1 - p_hat) * (1/n_A + 1/n_B))

# Estatística Z
z_obs = (p_B - p_A) / ep

# p-valor bicaudal
p_valor = 2 * stats.norm.sf(abs(z_obs))

# Z crítico
z_crit = stats.norm.ppf(1 - alpha / 2)

# Uplift relativo
uplift = (p_B - p_A) / p_A * 100

# Poder do teste (simulado)
poder_sim = np.mean([
    abs((np.random.binomial(n_B, p_B)/n_B -
         np.random.binomial(n_A, p_A)/n_A) /
        np.sqrt(p_hat*(1-p_hat)*(1/n_A+1/n_B))) > z_crit
    for _ in range(10_000)
])

# ══════════════════════════════════════════════════════════
# Console
# ══════════════════════════════════════════════════════════
print("=" * 58)
print("   TESTE A/B — BOTÃO DE COMPRA")
print("=" * 58)
print(f"   {'':20} {'Versão A':>10} {'Versão B':>10}")
print(f"   {'Usuários':20} {n_A:>10,} {n_B:>10,}")
print(f"   {'Conversões':20} {x_A:>10,} {x_B:>10,}")
print(f"   {'Taxa de conversão':20} {p_A:>9.2%} {p_B:>9.2%}")
print("-" * 58)
print(f"   Proporção combinada (p̂):  {p_hat:.4f}")
print(f"   Erro padrão (EP):         {ep:.4f}")
print(f"   Estatística Z:            {z_obs:.4f}")
print(f"   Z crítico (±):            {z_crit:.4f}")
print(f"   p-valor:                  {p_valor:.4f}")
print(f"   Uplift relativo:          +{uplift:.1f}%")
print(f"   Poder do teste:           {poder_sim:.2%}")
print("-" * 58)
decisao = "REJEITA H₀" if p_valor < alpha else "NÃO REJEITA H₀"
print(f"   Decisão ({alpha=}):       {decisao}")
if p_valor < alpha:
    print(f"   → Versão B é estatisticamente superior")
    print(f"   → Uplift de +{uplift:.1f}% na conversão")
print("=" * 58)

# ══════════════════════════════════════════════════════════
# GRÁFICOS
# ══════════════════════════════════════════════════════════
fig = plt.figure(figsize=(18, 18))
gs  = gridspec.GridSpec(3, 2, figure=fig,
                        hspace=0.50, wspace=0.35)

c_a    = '#1565C0'
c_b    = '#2E7D32'
c_verm = '#C62828'
c_larn = '#E65100'
c_cinz = '#546E7A'

# ══════════════════════════════════════════════════════════
# Painel 1 — Taxas de conversão com IC
# ══════════════════════════════════════════════════════════
ax1 = fig.add_subplot(gs[0, 0])

# IC para cada proporção
z95   = stats.norm.ppf(0.975)
ic_a  = z95 * np.sqrt(p_A * (1 - p_A) / n_A)
ic_b  = z95 * np.sqrt(p_B * (1 - p_B) / n_B)

barras = ax1.bar(['Versão A\n(Controle)', 'Versão B\n(Tratamento)'],
                 [p_A * 100, p_B * 100],
                 color=[c_a, c_b], alpha=0.75,
                 edgecolor='white', width=0.5)

ax1.errorbar(['Versão A\n(Controle)', 'Versão B\n(Tratamento)'],
             [p_A * 100, p_B * 100],
             yerr=[ic_a * 100, ic_b * 100],
             fmt='none', color='black',
             capsize=8, linewidth=2, capthick=2)

# Valores nas barras
for barra, val, ic in zip(barras, [p_A, p_B], [ic_a, ic_b]):
    ax1.text(barra.get_x() + barra.get_width()/2,
             val * 100 + ic * 100 + 0.2,
             f'{val:.1%} ± {ic:.2%}',
             ha='center', fontsize=10, fontweight='bold')

# Anotação do uplift
ax1.annotate('', xy=(1, p_B * 100), xytext=(1, p_A * 100),
             arrowprops=dict(arrowstyle='<->',
                             color=c_verm, lw=2))
ax1.text(1.28, (p_A + p_B) / 2 * 100,
         f'+{uplift:.1f}%\nuplift',
         ha='center', fontsize=10,
         color=c_verm, fontweight='bold')

ax1.set_title('Taxa de Conversão por Versão\ncom Intervalo de Confiança 95%',
              fontweight='bold', fontsize=11)
ax1.set_ylabel('Taxa de conversão (%)', fontsize=10)
ax1.set_ylim(0, p_B * 100 + 3)
ax1.spines[['top', 'right']].set_visible(False)

# ══════════════════════════════════════════════════════════
# Painel 2 — Distribuição Z com p-valor
# ══════════════════════════════════════════════════════════
ax2 = fig.add_subplot(gs[0, 1])

x_z = np.linspace(-4.5, 4.5, 600)
y_z = stats.norm.pdf(x_z)

ax2.plot(x_z, y_z, color=c_cinz, linewidth=2.5,
         label='Normal padrão (H₀)')

# Região crítica
x_rc_d = x_z[x_z >=  z_crit]
x_rc_e = x_z[x_z <= -z_crit]
ax2.fill_between(x_rc_d, stats.norm.pdf(x_rc_d),
                 alpha=0.45, color=c_verm,
                 label=f'Região crítica (α={alpha})')
ax2.fill_between(x_rc_e, stats.norm.pdf(x_rc_e),
                 alpha=0.45, color=c_verm)

# p-valor
x_pv_d = x_z[x_z >=  abs(z_obs)]
x_pv_e = x_z[x_z <= -abs(z_obs)]
ax2.fill_between(x_pv_d, stats.norm.pdf(x_pv_d),
                 alpha=0.65, color=c_larn,
                 label=f'p-valor = {p_valor:.4f}')
ax2.fill_between(x_pv_e, stats.norm.pdf(x_pv_e),
                 alpha=0.65, color=c_larn)

ax2.axvline( z_obs, color=c_b, linewidth=2.5,
            linestyle='--',
            label=f'Z obs = {z_obs:.3f}')
ax2.axvline( z_crit, color=c_verm, linewidth=1.8,
            linestyle=':', alpha=0.8,
            label=f'Z crítico = ±{z_crit:.3f}')
ax2.axvline(-z_crit, color=c_verm, linewidth=1.8,
            linestyle=':', alpha=0.8)

decisao_txt = 'REJEITA H₀ ✓' if p_valor < alpha else 'NÃO REJEITA H₀'
cor_dec     = c_verm if p_valor < alpha else c_b
ax2.text(0, stats.norm.pdf(0) * 0.4, decisao_txt,
         ha='center', fontsize=11, fontweight='bold',
         color=cor_dec,
         bbox=dict(boxstyle='round,pad=0.3',
                   facecolor='white',
                   edgecolor=cor_dec, alpha=0.9))

ax2.set_title('Estatística Z e p-valor\nTeste bicaudal H₀: pA = pB',
              fontweight='bold', fontsize=11)
ax2.set_xlabel('Estatística Z', fontsize=10)
ax2.set_ylabel('Densidade', fontsize=10)
ax2.legend(fontsize=8)
ax2.spines[['top', 'right']].set_visible(False)

# ══════════════════════════════════════════════════════════
# Painel 3 — Distribuição da diferença de proporções
# ══════════════════════════════════════════════════════════
ax3 = fig.add_subplot(gs[1, 0])

diff_obs  = p_B - p_A
ep_diff   = np.sqrt(p_A*(1-p_A)/n_A + p_B*(1-p_B)/n_B)
x_diff    = np.linspace(-0.05, 0.10, 500)

# Sob H₀ (centrada em 0)
y_H0_diff = stats.norm.pdf(x_diff, 0, ep)
ax3.plot(x_diff, y_H0_diff, color=c_cinz, linewidth=2.5,
         linestyle='--', label='Sob H₀ (diff=0)',
         alpha=0.7)

# Sob H₁ (centrada na diferença observada)
y_H1_diff = stats.norm.pdf(x_diff, diff_obs, ep_diff)
ax3.plot(x_diff, y_H1_diff, color=c_b, linewidth=2.5,
         label=f'Sob H₁ (diff={diff_obs:.2%})')

# Região crítica na escala da diferença
lim_d = z_crit * ep
ax3.axvline( lim_d, color=c_verm, linewidth=2,
            linestyle=':', label=f'Limite crítico = {lim_d:.4f}')
ax3.axvline(-lim_d, color=c_verm, linewidth=2, linestyle=':')
ax3.axvline(diff_obs, color=c_larn, linewidth=2.5,
            linestyle='--',
            label=f'Diff obs = {diff_obs:.2%}')
ax3.axvline(0, color=c_cinz, linewidth=1.2,
            linestyle='-', alpha=0.5, label='H₀: diff = 0')

# Área p-valor
x_pv_diff = x_diff[x_diff >= diff_obs]
ax3.fill_between(x_pv_diff, stats.norm.pdf(x_pv_diff, 0, ep),
                 alpha=0.4, color=c_larn)
x_pv_diff2 = x_diff[x_diff <= -diff_obs]
ax3.fill_between(x_pv_diff2, stats.norm.pdf(x_pv_diff2, 0, ep),
                 alpha=0.4, color=c_larn)

ax3.set_title('Distribuição da Diferença pB − pA\n'
              'H₀: diferença = 0  vs  H₁: diferença ≠ 0',
              fontweight='bold', fontsize=11)
ax3.set_xlabel('Diferença de proporções (pB − pA)', fontsize=10)
ax3.set_ylabel('Densidade', fontsize=10)
ax3.legend(fontsize=8)
ax3.spines[['top', 'right']].set_visible(False)

# ══════════════════════════════════════════════════════════
# Painel 4 — Simulação: distribuição de diferenças sob H₀
# ══════════════════════════════════════════════════════════
ax4 = fig.add_subplot(gs[1, 1])

N_sim     = 50_000
diffs_sim = np.array([
    np.random.binomial(n_B, p_hat)/n_B -
    np.random.binomial(n_A, p_hat)/n_A
    for _ in range(N_sim)
])

ax4.hist(diffs_sim, bins=80, density=True,
         color=c_cinz, alpha=0.6, edgecolor='none',
         label=f'Simulação (N={N_sim:,})')

x_sim = np.linspace(diffs_sim.min(), diffs_sim.max(), 400)
ax4.plot(x_sim, stats.norm.pdf(x_sim, 0, ep),
         color='black', linewidth=2, label='Normal teórica')

# Diferença observada
ax4.axvline( diff_obs, color=c_larn, linewidth=2.5,
            linestyle='--',
            label=f'Diff obs = {diff_obs:.2%}')
ax4.axvline(-diff_obs, color=c_larn, linewidth=2.5,
            linestyle='--')

# p-valor empírico
p_emp = np.mean(np.abs(diffs_sim) >= abs(diff_obs))
ax4.fill_between(x_sim[x_sim >= diff_obs],
                 stats.norm.pdf(x_sim[x_sim >= diff_obs], 0, ep),
                 alpha=0.5, color=c_larn,
                 label=f'p empírico = {p_emp:.4f}')
ax4.fill_between(x_sim[x_sim <= -diff_obs],
                 stats.norm.pdf(x_sim[x_sim <= -diff_obs], 0, ep),
                 alpha=0.5, color=c_larn)

ax4.set_title(f'Simulação sob H₀ (N={N_sim:,})\n'
              f'p-valor teórico={p_valor:.4f}  '
              f'p-valor empírico={p_emp:.4f}',
              fontweight='bold', fontsize=11)
ax4.set_xlabel('Diferença simulada (pB − pA)', fontsize=10)
ax4.set_ylabel('Densidade', fontsize=10)
ax4.legend(fontsize=8)
ax4.spines[['top', 'right']].set_visible(False)

# ══════════════════════════════════════════════════════════
# Painel 5 — Análise de poder: n necessário
# ══════════════════════════════════════════════════════════
ax5 = fig.add_subplot(gs[2, 0])

ns       = np.arange(500, 8001, 100)
poderes  = []
efeito   = abs(p_B - p_A)

for n_i in ns:
    ep_i    = np.sqrt(2 * p_hat * (1 - p_hat) / n_i)
    z_beta  = (efeito / ep_i) - z_crit
    poder_i = stats.norm.cdf(z_beta)
    poderes.append(poder_i)

poderes = np.array(poderes)

ax5.plot(ns, poderes * 100, color=c_b,
         linewidth=2.5, label='Poder do teste')
ax5.axhline(80, color=c_larn, linewidth=2,
            linestyle='--', label='Poder mínimo recomendado (80%)')
ax5.axhline(95, color=c_verm, linewidth=1.5,
            linestyle=':', label='Poder ideal (95%)')
ax5.axvline(n_A, color=c_cinz, linewidth=2,
            linestyle='-.', label=f'n usado = {n_A:,}')

n_80 = ns[np.argmax(poderes >= 0.80)]
n_95 = ns[np.argmax(poderes >= 0.95)]
ax5.axvline(n_80, color=c_larn, linewidth=1.5,
            linestyle=':',
            label=f'n para 80% = {n_80:,}')
ax5.axvline(n_95, color=c_verm, linewidth=1.5,
            linestyle=':',
            label=f'n para 95% = {n_95:,}')

ax5.fill_between(ns, poderes * 100, 80,
                 where=poderes * 100 >= 80,
                 alpha=0.15, color=c_b,
                 label='Zona de poder adequado')

ax5.set_title('Poder do Teste vs Tamanho da Amostra\n'
              f'Efeito detectado: {efeito:.2%}  α={alpha}',
              fontweight='bold', fontsize=11)
ax5.set_xlabel('Tamanho da amostra por grupo (n)', fontsize=10)
ax5.set_ylabel('Poder do teste (%)', fontsize=10)
ax5.set_ylim(0, 105)
ax5.legend(fontsize=8)
ax5.spines[['top', 'right']].set_visible(False)

# ══════════════════════════════════════════════════════════
# Painel 6 — Impacto de negócio
# ══════════════════════════════════════════════════════════
ax6 = fig.add_subplot(gs[2, 1])
ax6.axis('off')

ticket_medio   = 250
usuarios_mes   = 50_000
conv_a_mes     = int(usuarios_mes * p_A)
conv_b_mes     = int(usuarios_mes * p_B)
receita_a      = conv_a_mes * ticket_medio
receita_b      = conv_b_mes * ticket_medio
ganho_mes      = receita_b - receita_a
ganho_ano      = ganho_mes * 12

linhas = [
    ('RESULTADO DO TESTE A/B', '', True),
    ('', '', False),
    ('Métrica',          'Versão A',               False),
    ('Usuários',         f'{n_A:,}',               False),
    ('Conversões',       f'{x_A:,}',               False),
    ('Taxa',             f'{p_A:.1%}',             False),
    ('', '', False),
    ('Métrica',          'Versão B',               False),
    ('Usuários',         f'{n_B:,}',               False),
    ('Conversões',       f'{x_B:,}',               False),
    ('Taxa',             f'{p_B:.1%}',             False),
    ('', '', False),
    ('ESTATÍSTICA',      '',                       True),
    ('Z observado',      f'{z_obs:.4f}',           False),
    ('Z crítico',        f'±{z_crit:.4f}',         False),
    ('p-valor',          f'{p_valor:.4f}',         False),
    ('Decisão',          'REJEITA H₀ ✓',          False),
    ('', '', False),
    ('IMPACTO NEGÓCIO',  '',                       True),
    ('Ticket médio',     f'R$ {ticket_medio:,}',   False),
    ('Usuários/mês',     f'{usuarios_mes:,}',      False),
    ('Receita A/mês',    f'R$ {receita_a:,}',      False),
    ('Receita B/mês',    f'R$ {receita_b:,}',      False),
    ('Ganho/mês',        f'R$ {ganho_mes:,}',      False),
    ('Ganho/ano',        f'R$ {ganho_ano:,}',      False),
]

y = 0.98
for col1, col2, negrito in linhas:
    if col1 == '' and col2 == '':
        y -= 0.025
        continue
    peso = 'bold' if negrito else 'normal'
    cor  = c_b    if negrito else '#333333'
    ax6.text(0.02, y, col1, transform=ax6.transAxes,
             fontsize=9.5, fontweight=peso, color=cor,
             va='top')
    if col2:
        cor2 = c_verm if 'REJEITA' in col2 else '#333333'
        ax6.text(0.62, y, col2, transform=ax6.transAxes,
                 fontsize=9.5, fontweight=peso,
                 color=cor2, va='top')
    y -= 0.038

ax6.set_title('Resumo Executivo',
              fontweight='bold', fontsize=11, pad=10)

plt.suptitle(
    f'Teste A/B — Botão de Compra  |  '
    f'pA={p_A:.1%}  pB={p_B:.1%}  '
    f'Uplift=+{uplift:.1f}%  '
    f'p-valor={p_valor:.4f}  →  {"Versão B Vence ✓" if p_valor < alpha else "Sem diferença significativa"}',
    fontsize=12, fontweight='bold', y=1.01)

plt.tight_layout()
plt.show()
````
